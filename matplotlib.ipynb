```python
# üîë Core Components

# matplotlib ‚Üí The main library for data visualization in Python.

# pyplot ‚Üí A module inside matplotlib (commonly imported as plt) for making plots easily.

# Figure ‚Üí The entire window or page that holds your plots.

# Axes ‚Üí The actual plotting area inside a Figure (think of it as one graph).

# Axis ‚Üí The x and y lines with ticks/labels inside Axes.

# üìä Plotting Functions

# plot() ‚Üí Line plot.

# scatter() ‚Üí Scatter plot.

# bar() / barh() ‚Üí Vertical/Horizontal bar chart.

# hist() ‚Üí Histogram.

# pie() ‚Üí Pie chart.

# imshow() ‚Üí Display images (matrix/2D array).

# üé® Customization

# Labeling

# xlabel() ‚Üí X-axis label.

# ylabel() ‚Üí Y-axis label.

# title() ‚Üí Title of the plot.

# legend() ‚Üí Add a legend for multiple plots.

# text() ‚Üí Add custom text at a position.

# Ticks

# xticks() / yticks() ‚Üí Set tick locations & labels.

# Grid

# grid() ‚Üí Add grid lines.

# üñå Styling

# Color (c) ‚Üí e.g., "red", "#ff5733", or shorthand "r".

# Line style (ls) ‚Üí "-", "--", "-.", ":".

# Marker (marker) ‚Üí "o", "s", "^", "*".

# Alpha ‚Üí Transparency (0 = invisible, 1 = solid).

# Label ‚Üí For legends.
# ‚öôÔ∏è Figure Management

# figure() ‚Üí Create a new figure.

# subplots() ‚Üí Multiple plots in one figure.

# subplot() ‚Üí Position plots in grid layout.

# tight_layout() ‚Üí Adjust spacing automatically.

# savefig() ‚Üí Save figure to file (PNG, JPG, PDF, etc.).

# show() ‚Üí Display the figure.
# üìê Advanced

# Object-Oriented API

# fig, ax = plt.subplots() ‚Üí Create Figure and Axes objects.

# ax.plot(), ax.set_title(), etc. ‚Üí More control over plots.

# Colormap (cmap) ‚Üí Color schemes for heatmaps, scatter, images.

# Annotations ‚Üí annotate() to mark points with arrows/notes.

```

```python
#plt.plot(x,y color="color name",linestyle ="linestyle",linewidth=value ,marker="markersymbol",label="labelname")
import matplotlib.pyplot as plt
months =[1,2,3,4]
sales =[1000,2000,1500,2500]
plt.plot(months,sales,color="red",linestyle="--",linewidth=2,marker="o",label="2025 sales data")
plt.xlabel("months")
plt.ylabel("sales per month")
plt.title("total sales")
plt.grid(color="blue",linestyle=":",linewidth=1)
plt.legend(loc="upper left",fontsize =6)#bottom right
plt.xlim(1,4)
plt.ylim(1000,3000)
plt.xticks([1,2,3,4],["M1","M2","M3","M4"])
plt.show()
```

```python
#plt.pie(values,labels =label_list,colors=color_list,autopct='%1.1f%%')   autopct='%1.1f%%'<-this is for showing percent in a bar
regions =["North","South","East","West"]
revenue=[3000,4400,220,1122]
plt.pie(revenue,labels=regions,autopct='%1.1f%%',colors=["gold","coral","skyblue","lightpink"])
plt.title("revenue contribution by region")
plt.show()
```

```python
# To make the East slice stand out (a little outward from the pie chart), you can use the explode parameter in plt.pie().
import matplotlib.pyplot as plt

regions = ["North", "South", "East", "West"]
revenue = [3000, 4400, 220, 1122]

# Explode only the "East" slice (3rd element, index=2)
explode = [0, 0, 0.9, 0]  

plt.pie(revenue,
        labels=regions,
        autopct='%1.1f%%',
        colors=["gold", "coral", "skyblue", "lightpink"],
        explode=explode,
        shadow=True,startangle=50)

plt.title("Revenue Contribution by Region")
plt.show()
```

subplot and layout adjustments

```python
#  Subplots

# plt.subplot(rows, cols, index) ‚Üí Creates a grid of plots in one figure.
# Example: plt.subplot(2,2,1) ‚Üí 2 rows √ó 2 cols grid, first plot.

# plt.subplots(rows, cols) ‚Üí Returns a figure and an array of axes (more flexible).

#  Used when you want multiple charts in one window.

#  Layout Adjustments

# plt.tight_layout() ‚Üí Automatically adjusts spacing so labels/titles don‚Äôt overlap.

# plt.subplots_adjust(left, right, top, bottom, wspace, hspace)

# wspace ‚Üí width space between plots.

# hspace ‚Üí height space between plots.

#  Used when subplots overlap or need better spacing.

#  In short:

# Subplots ‚Üí arrange multiple plots in a grid.

# Layout adjustments ‚Üí fix spacing issues for clarity.
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[10,22,50,42]
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("line chart")

plt.subplot(1,2,2)
plt.bar(x,y)
plt.title("barchart")
plt.show()
```

```python
import matplotlib.pyplot as plt


#fig,axis =plt.subplots(nrows,ncols,figsize=(width,height))
fig,ax =plt.subplots(1,2,figsize =(8,5))
x=[1,2,3,4]
y=[10,22,50,42]
ax[0].plot(x,y,color="purple")
ax[0].set_title("line plot")

ax[1].bar(x,y,color="purple")
ax[1].set_title("bar plot")
fig.suptitle("comparision b/w line and bar chart")
plt.tight_layout()
plt.show()

```
