```python
# 🔑 Core Components

# matplotlib → The main library for data visualization in Python.

# pyplot → A module inside matplotlib (commonly imported as plt) for making plots easily.

# Figure → The entire window or page that holds your plots.

# Axes → The actual plotting area inside a Figure (think of it as one graph).

# Axis → The x and y lines with ticks/labels inside Axes.

# 📊 Plotting Functions

# plot() → Line plot.

# scatter() → Scatter plot.

# bar() / barh() → Vertical/Horizontal bar chart.

# hist() → Histogram.

# pie() → Pie chart.

# imshow() → Display images (matrix/2D array).

# 🎨 Customization

# Labeling

# xlabel() → X-axis label.

# ylabel() → Y-axis label.

# title() → Title of the plot.

# legend() → Add a legend for multiple plots.

# text() → Add custom text at a position.

# Ticks

# xticks() / yticks() → Set tick locations & labels.

# Grid

# grid() → Add grid lines.

# 🖌 Styling

# Color (c) → e.g., "red", "#ff5733", or shorthand "r".

# Line style (ls) → "-", "--", "-.", ":".

# Marker (marker) → "o", "s", "^", "*".

# Alpha → Transparency (0 = invisible, 1 = solid).

# Label → For legends.
# ⚙️ Figure Management

# figure() → Create a new figure.

# subplots() → Multiple plots in one figure.

# subplot() → Position plots in grid layout.

# tight_layout() → Adjust spacing automatically.

# savefig() → Save figure to file (PNG, JPG, PDF, etc.).

# show() → Display the figure.
# 📐 Advanced

# Object-Oriented API

# fig, ax = plt.subplots() → Create Figure and Axes objects.

# ax.plot(), ax.set_title(), etc. → More control over plots.

# Colormap (cmap) → Color schemes for heatmaps, scatter, images.

# Annotations → annotate() to mark points with arrows/notes.

```

```python
# 🔥 Pyplot vs Object-Oriented API in Matplotlib
# 1️⃣ Style of Coding
# Pyplot API (State-based)
# import matplotlib.pyplot as plt
# import numpy as np

# x = np.linspace(0, 10, 100)
# y = np.sin(x)

# plt.plot(x, y)             # directly plot
# plt.title("Sine Wave")     # set title
# plt.xlabel("X-axis")
# plt.ylabel("Y-axis")
# plt.show()


# Works like MATLAB.

# State-based: functions act on the “current figure/axes”.

# Quick for simple plots.

# Object-Oriented API (Figure & Axes-based)
# fig, ax = plt.subplots()

# ax.plot(x, y)              # plot using Axes object
# ax.set_title("Sine Wave")  
# ax.set_xlabel("X-axis")
# ax.set_ylabel("Y-axis")

# plt.show()


# Explicitly creates Figure & Axes objects.

# Functions are methods of Axes.

# More structured & scalable.

# 2️⃣ Multiple Plots
# Pyplot (harder to manage)
# plt.subplot(1,2,1)
# plt.plot(x, np.sin(x))
# plt.title("Sin")

# plt.subplot(1,2,2)
# plt.plot(x, np.cos(x))
# plt.title("Cos")

# plt.show()


# Uses plt.subplot() with grid indexing.

# Gets confusing for many subplots.

# OO API (cleaner & flexible)
# fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8,4))

# ax1.plot(x, np.sin(x))
# ax1.set_title("Sin")

# ax2.plot(x, np.cos(x))
# ax2.set_title("Cos")

# plt.show()


# Each subplot is an Axes object (ax1, ax2).

# Easy to customize each one.

# 3️⃣ Customization
# Pyplot
# plt.plot(x, y)
# plt.xlim(0, 10)
# plt.ylim(-1.5, 1.5)

# OO API
# fig, ax = plt.subplots()
# ax.plot(x, y)
# ax.set_xlim(0, 10)
# ax.set_ylim(-1.5, 1.5)


# Pyplot → uses global state.

# OO API → customizes via methods of Axes.

# 4️⃣ When to Use

# ✅ Pyplot API

# Quick, small scripts.

# Exploratory data analysis (EDA).

# One-off plots in Jupyter.

# ✅ Object-Oriented API

# Multiple subplots/figures.

# Dashboards or reports.

# Complex visualizations with fine control.

# Professional & production code.

# 5️⃣ Summary Table
# Feature-	Pyplot API	Object-Oriented API
# Style-	State-based (MATLAB-like) \	Explicit (Figure & Axes objects)
# Ease of use-	Very easy for quick plots \	More verbose but structured
# Best for-	Simple, small scripts \	Complex, multiple subplots/figures
# Customization-	Limited, can get messy \	Fine-grained control
# Scalability-	Poor for many plots \	Excellent for large projects
# Professional use-	Rare \	Standard practice
```

```python
#plt.plot(x,y color="color name",linestyle ="linestyle",linewidth=value ,marker="markersymbol",label="labelname")
import matplotlib.pyplot as plt
months =[1,2,3,4]
sales =[1000,2000,1500,2500]
plt.plot(months,sales,color="red",linestyle="--",linewidth=2,marker="o",label="2025 sales data")
plt.xlabel("months")
plt.ylabel("sales per month")
plt.title("total sales")
plt.grid(color="blue",linestyle=":",linewidth=1)
plt.legend(loc="upper left",fontsize =6)#bottom right
plt.xlim(1,4)
plt.ylim(1000,3000)
plt.xticks([1,2,3,4],["M1","M2","M3","M4"])
plt.show()
```

```python
#plt.pie(values,labels =label_list,colors=color_list,autopct='%1.1f%%')   autopct='%1.1f%%'<-this is for showing percent in a bar
regions =["North","South","East","West"]
revenue=[3000,4400,220,1122]
plt.pie(revenue,labels=regions,autopct='%1.1f%%',colors=["gold","coral","skyblue","lightpink"])
plt.title("revenue contribution by region")
plt.show()
```

```python
# To make the East slice stand out (a little outward from the pie chart), you can use the explode parameter in plt.pie().
import matplotlib.pyplot as plt

regions = ["North", "South", "East", "West"]
revenue = [3000, 4400, 220, 1122]

# Explode only the "East" slice (3rd element, index=2)
explode = [0, 0, 0.9, 0]  

plt.pie(revenue,
        labels=regions,
        autopct='%1.1f%%',
        colors=["gold", "coral", "skyblue", "lightpink"],
        explode=explode,
        shadow=True,startangle=50)

plt.title("Revenue Contribution by Region")
plt.show()
```

subplot and layout adjustments

```python
#  Subplots

# plt.subplot(rows, cols, index) → Creates a grid of plots in one figure.
# Example: plt.subplot(2,2,1) → 2 rows × 2 cols grid, first plot.

# plt.subplots(rows, cols) → Returns a figure and an array of axes (more flexible).

#  Used when you want multiple charts in one window.

#  Layout Adjustments

# plt.tight_layout() → Automatically adjusts spacing so labels/titles don’t overlap.

# plt.subplots_adjust(left, right, top, bottom, wspace, hspace)

# wspace → width space between plots.

# hspace → height space between plots.

#  Used when subplots overlap or need better spacing.

#  In short:

# Subplots → arrange multiple plots in a grid.

# Layout adjustments → fix spacing issues for clarity.
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[10,22,50,42]
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("line chart")

plt.subplot(1,2,2)
plt.bar(x,y)
plt.title("barchart")
plt.show()
```

```python
import matplotlib.pyplot as plt


#fig,axis =plt.subplots(nrows,ncols,figsize=(width,height))
fig,ax =plt.subplots(1,2,figsize =(8,5))
x=[1,2,3,4]
y=[10,22,50,42]
ax[0].plot(x,y,color="purple")
ax[0].set_title("line plot")

ax[1].bar(x,y,color="purple")
ax[1].set_title("bar plot")
fig.suptitle("comparision b/w line and bar chart")
plt.tight_layout()
plt.show()

```
